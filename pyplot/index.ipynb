### Data visualization using Pyplot: pie chart, line chart, and bar chart.

``` PIE CHART ```

````python
import matplotlib.pyplot as plt
sub = 'Arts','Science','Commerce'
per = [78,95,86]
explode = (0,0.1,0)
plt.pie(per,explode=explode,labels=sub,autopct='%1.2f%%',shadow=True)
plt.axis('equal')
plt.title('Percentage Graph')
plt.show()
````

``` LINE CHART ```

````python
import matplotlib.pyplot as plt
x = ['JavaScript','PHP','Python','C#','Java']
y = [92,90,96,79,88]
plt.plot(x,y)
plt.xlabel('Languages')
plt.ylabel('Popularity Score')
plt.title('Language Graph')
plt.show()
````

``` BAR CHART ```

````python
import matplotlib.pyplot as plt
import numpy as np
x = ['AS','KA','KL','MP','MH','TN','UP','UK','WB']
y = [190,524,190,526,312,264,173,442,88]
ind = np.arange(len(x))
# ind
plt.title('Tiger population IN. 2011')
plt.bar(ind,y)
plt.xticks(ind,x)
plt.show()
````

``` LINE CHART COMPARISON ```

````python
import matplotlib.pyplot as plt
sub = ['CS','IT','ME','EEE']
s1 = [96,94,85,89]
s2 = [92,89,78,93]
plt.plot(sub,s1,label="Jerry")
plt.plot(sub,s2,label="Tom")
plt.xlabel = ('Streams')
plt.ylabel = ('Scores')
plt.title = ('Comparison Graph')
plt.legend()
plt.show()
````
